```python
data_dir = '/data2/object-masking/datasets/mb1_base_may_80_dataset/data'
mask_dir = '/data2/object-masking/datasets/mb1_base_may_80_dataset/mask'
im_shape = None
name = '01121557'
```

```python
import os
from keras.callbacks import ModelCheckpoint
from src.models.unet_mb import get_unet, load_data

X, Y, mean_and_std = load_data(data_dir, mask_dir, im_shape)

in_shape = X[0].shape
model = get_unet(in_shape)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

if not os.path.isdir('tmps'):
    os.system('mkdir -m 777 -p {}'.format('tmps'))

checkpoint = ModelCheckpoint(filepath='tmps/{}.unet.checkpoint'.format(name), verbose=0)
best_model = ModelCheckpoint(filepath='tmps/{}.unet.best.model'.format(name), verbose=0, save_best_only=True)

model.fit(X, Y, batch_size=16, epochs=900, callbacks=[checkpoint, best_model], validation_split=0.2)
model.evaluate(X, Y, batch_size=16, verbose=0)
```
